## Clean on Job Title Column

```python
data_Glassdor['Job Title'].value_counts()
```

```python
data_Glassdor['Job Title'].unique()
```

```python
def job_title_classifier(title):
    
    if 'data scientist' in title.lower():
        return 'Data scientist'
    elif 'data engineer' in title.lower():
        return 'Data engineer'
    elif 'analyst' in title.lower():
        return 'Analyst'
    elif 'machine learning' in title.lower():
        return 'MLE'
    elif 'manager' in title.lower():
        return 'Manager'
    elif 'director' in title.lower():
        return 'Director'
    else:
        return 'N/A'
```

```python
data_Glassdor['Job_Simp'] = data_Glassdor['Job Title'].apply(job_title_classifier)
data_Glassdor['Job_Simp'].value_counts()
```

```python
data_Glassdor.head(2)
```

## Job Level Classifier

```python
def job_level_classifier(title):
    if 'sr' in title.lower() or 'senior' in title.lower() or 'sr' in title.lower() or 'lead' in title.lower() or 'principal' in title.lower():
        return 'Senior'
    elif 'jr' in title.lower() or 'jr.' in title.lower():
        return 'Junior'
    else:
        return 'NA'
```

```python
data_Glassdor['Seniority'] = data_Glassdor['Job Title'].apply(job_level_classifier)
data_Glassdor['Seniority'].value_counts()
```

```python
data_Glassdor.head(2)
```

## Clean on Salary Estimate column

```python
data_Glassdor['Salary Estimate'].value_counts()
```

```python
data_Glassdor['Salary Estimate'].unique()
```

```python
data_Glassdor['Salary Estimate'] = data_Glassdor['Salary Estimate'].apply(lambda x: x.split('(')[0])
```

```python
data_Glassdor['Salary Estimate'].unique()
```

```python
data_Glassdor['Salary Estimate'] = data_Glassdor['Salary Estimate'].apply(lambda x: x.replace('K','').replace('$',''))
```

```python
data_Glassdor['Salary Estimate'].unique()
```

```python
data_Glassdor['min_salary'] = data_Glassdor['Salary Estimate'].apply(lambda x: int(x.split('-')[0]))

data_Glassdor['max_salary'] = data_Glassdor['Salary Estimate'].apply(lambda x: int(x.split('-')[1]))
```

```python
data_Glassdor['avg_Salary'] = (data_Glassdor['min_salary']+ data_Glassdor['max_salary'])//2
```

```python
data_Glassdor.head(2)
```
