# Work on the Contract Column

```python
fifa['Contract'].dtype
```

```python
fifa['Contract'].unique()
```

```python
for index, row in fifa.iterrows():
    if 'On Loan' in row ['Contract'] or 'Free' in row ['Contract']:
        print(row['Contract'])
```

```python
def extract_contract_info(contract):
    if contract == 'Free' or 'On Loan' in contract:
        start_date = np.nan
        end_date = np.nan
        contract_length = 0
    else:
        start_date, end_date = contract.split (' ~ ')
        start_year = int(start_date [ : 4])
        end_year = int(end_date [ : 4])
        contract_length = end_year - start_year
    return start_date, end_date, contract_length

# Apply function to Contract Column and Create new Columns
new_cols = ['Contract Start', 'Contract End', 'Contract Length(years)']
new_data = fifa['Contract'].apply(lambda x : pd.Series(extract_contract_info(x)))

for i in range(len(new_cols)):
    fifa.insert(loc=fifa.columns.get_loc('Contract')+1+i, column = new_cols[i], value= new_data[i])
```

```python
fifa.head()
```

```python
#Contract Status
def categorize_contract_status(contract):
    if contract == 'Free':
        return 'Free'
    elif 'On Loan' in contract:
        return 'On Loan'
    else:
        return 'Contract'

# Add Contract Status Column
fifa.insert(fifa.columns.get_loc('Contract Length(years)')+1, 'Contract Status', fifa['Contract'].apply(categorize_contract_status))
fifa.head(3)
```

# Change the height and weight column

```python
fifa['Height'].dtype
```

```python
fifa['Height'].unique()
```

```python
def convert_height(x):
    if x[-1] == '"':
        x = x.replace("\"","")
        inch = int (x[2:]) * 2.54
        foot = int(x[0]) * 30.48
        return round(foot+inch)
    elif x[-1] == "m":
        return int(x[:2])
```

```python
fifa.Height = fifa.Height.apply(convert_height)
fifa.Height.unique()
```

```python
# rename the column name Height to Height_cm
fifa = fifa.rename(columns = {'Height': 'Height_cm'})
```

```python
fifa.head(2)
```

```python
fifa['Weight'].dtype
```

```python
fifa['Weight'].unique()
```

```python
def convert_weight(weight):
    if "kg" in weight:
        return int(weight.strip("kg"))
    else:
        pounds = int(weight.strip("lbs"))
        return round(pounds/2/205)

# Apply Function to Weight column
fifa['Weight'] = fifa['Weight'].apply(convert_weight)
fifa['Weight'].unique()
```

```python
# Rename the column name Weight to Weight(kg)
fifa = fifa.rename(columns ={'Weight':'Weight(kg)'})
fifa.head(2)
```

## RELEASE CLAUSE

```python
# Release Clause
fifa['Release Clause'].unique()
```

```python
def release(x):
    if 'M' in x:
        x = x.replace('M', '')
        x = float(x[1:]) * 1000000
        return round(x)
    elif "K" in x:
        x = x.replace('K', '')
        x = float(x[1:]) * 1000
        return round(x)
    else:
        return x
```

```python
fifa['Release Clause'] = fifa['Release Clause'].apply(release)
```

```python
# Rename the column to Release Clause to Release_Clause €
fifa = fifa.rename(columns = {'Release Clause':'Release_Clause €'})
```
